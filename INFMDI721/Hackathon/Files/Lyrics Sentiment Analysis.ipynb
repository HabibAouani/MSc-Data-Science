## Build models

### a. Create Features and label

```python
print("Loading dataset")

label_list = ['relaxed','angry','happy','sad']
path = '/Users/anatoli_debradke/Downloads/'

from glob import glob

filenames_angry = sorted(glob(op.join(path,'angry', '*.txt')))
filenames_sad = sorted(glob(op.join(path,'sad', '*.txt')))
filenames_happy = sorted(glob(op.join(path,'happy', '*.txt')))
filenames_relaxed = sorted(glob(op.join(path,'relaxed', '*.txt')))

texts_angry = [open(f).read() for f in filenames_angry]
texts_sad = [open(f).read() for f in filenames_sad]
texts_happy = [open(f).read() for f in filenames_happy]
texts_relaxed = [open(f).read() for f in filenames_relaxed]

texts = texts_angry + texts_sad + texts_happy + texts_relaxed

y = np.zeros(len(texts), dtype=np.int)
y[len(texts_angry) : len(texts_angry) + len(texts_sad)] = 1.0
y[len(texts_angry) + len(texts_sad) : len(texts_angry) + len(texts_sad) + len(texts_happy)] = 2.0
y[len(texts_angry) + len(texts_sad) + len(texts_happy):] = 3.0

stop_words_path = '/Users/anatoli_debradke/Desktop/MS_BGD/MDI343 - Machine Learning/TP/TP3/data/english.stop'
stop_word = open(stop_words_path).read()
stop_word = stop_word.split()
```

### b. Clean features

```python
for i in range(0, len(texts)) :
    texts[i] = texts[i].replace("\n", " ").replace("\'", " ")
    texts[i] = re.sub(r"\[(.*?)\]", " ", texts[i])
```

### c. Train and test dataset

```python
from sklearn.model_selection import KFold
# Spare l'chantillon train en 5 folds 
kfolds = KFold(n_splits=5, shuffle=True)

# Cross Validation du score
X_train = []
y_train = []
X_test = []
y_test = []

for idx_train, idx_test in kfolds.split(texts):
    XX = []
    yy = []
    for idx in idx_train:
        XX.append(texts[idx])
        yy.append(y[idx])
    X_train.append(XX)
    y_train.append(np.asarray(yy))
   
    XX = []
    yy = []
    for idx in idx_test:
        XX.append(texts[idx])
        yy.append(y[idx])
    X_test.append(XX)
    y_test.append(np.asarray(yy))
```

### d.Train model classification

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline

# Pipeline
pipeline = Pipeline([('vect', CountVectorizer()), ('lr', LogisticRegression())])

# Cross-validation
Score = []
for k in range(5):

    model = pipeline.fit(X_train[k], y_train[k])
    
    predict = model.predict(X_test[k])
    
    Score.append(np.mean(predict == y_test[k]))

print("Accuracy Score LinearRegression cross-validation (5 folds): {} %".format(round(np.mean(Score),3)))
```
