## Question 7 

Pour les 2 méthodes (OLS et PCA before OLS) : Tracer les résidus de la prédiction sur l’échantillon test. Tracer leur densité (on pourra par exemple utiliser un histogramme). Calculer le coefficient de détermination sur l’échantillon test. Calculer le risque de prédiction sur l’échantillon test.

```python
# Plotter les résidus PCA before OLS

# ACP

pca.fit(x_test)
x_train_pca = pca.transform(x_train)

# Applique les MCO sur ACP

reg_pca = LinearRegression().fit(x_train_pca, y_train)
ypred_pca_test = reg_pca.predict(pca.transform(x_test))

# MCO sans ACP 

reg_ols_test = LinearRegression().fit(x_train, y_train)
ypred_ols_test = reg_ols_test.predict(x_test)

resid_pca = (y_test - ypred_pca_test)
resid_ols = (y_test - ypred_ols_test)

# Graphique PCA

plt.figure()

plt.title('Les résidus avec ACP avant')
plt.scatter(range(0,len(resid_pca)), resid_pca)
plt.ylabel('Résidus')
plt.rcParams["figure.figsize"] = (10,7)

plt.show()

# Graphique OLS

plt.figure()

plt.title('Les moindres carrés sans ACP avant')
plt.scatter(range(0,len(resid_ols)), resid_ols)
plt.ylabel('Résidus')
plt.rcParams["figure.figsize"] = (10,7)

plt.show()

```

```python
# Densités 
plt.figure()

plt.title('Densités des résidus')
line1 = sns.distplot(resid_pca, kde=True, rug=True, label = 'Résidus PCA before OLS')
line2 = sns.distplot(resid_ols, kde=True, rug=True, label = 'Résidus OLS')
line1.legend()
line2.legend()
plt.rcParams["figure.figsize"] = (10,7)

plt.show()
```

```python
# Coefficient de déterminations 

print("R² pour PCA before OLS : {0}".format(reg_pca.score(pca.transform(x_test), y_test)))
print("R² pour OLS : {0}".format(reg_ols.score(x_test, y_test)))

# Risque de prédiction sur échantillon test : E((Y* - Yhat)²) / n

mse_ols = ((resid_ols)**2).mean()
mse_pca = ((resid_pca)**2).mean()

print("Risque de prédiction sur échantillon test OLS : {0}\nRisque de prédiction sur échantillon test PCA before OLS: {1}"
      .format(mse_ols, mse_pca))
```

## Question 8 

Appliquer la méthode de forward variable sélection. On pourra utiliser la statistique du test de nullité du coefficient (comme vu en cours). Pour l’instant, on ne met pas de critère d’arret sur la méthode. C’est à dire que l’on ajoute une variable à chaque étape jusqu’à retrouver la totalité des variables. Afficher l’ordre de séléction des variables.

```python
# Forward variable selection 

def forward_selection():
    features = []
    list_ttest = []
    list_pvalues = []
    y = y_train
    for i in range(210):
        features_model = list(set(range(210))- set(features))
        model = list(map(lambda x : sm.OLS(y, sm.add_constant(x_train.loc[:,x])).fit(), features_model))
        ttest = list(map(lambda x : abs(x.tvalues[1]), model))
        pvalues = list(map(lambda x : x.pvalues[1], model))
        index_max = ttest.index(max(ttest))
        list_ttest.append(ttest)
        list_pvalues.append(pvalues[index_max])
        features.append(features_model[index_max])
        y = y - model[index_max].predict(sm.add_constant(x_train.loc[:,features_model[index_max]]))
    df_t = pd.DataFrame(list_ttest)
    df_p = pd.DataFrame(list_pvalues)
    df_p ['x'] = pd.DataFrame(features)
    df_p.columns = ['pvalues', 'x']
    return features, df_t, df_p
        
features, df_t, df_p = forward_selection()
```

```python
print("Les x choisis :", features)
```

Afin de préparer la validation croisée, séparer l’échantillon train en 4 parties (appelées ”folds”) de façon aléatoire. On affichera les numéros d’échantillon sélectionnés dans chaque fold.

## Question 12

Appliquer la méthode de la régression ridge. Pour le choix du paramètre de régularisation, on fera une validation croisée sur les ”folds” définies lors de la question précédente. A tour de rôle chacune des ”folds” servira pour calculer le risque de prédiction alors que les autres seront utilisées pour estimer le modèle. On moyennera ensuite les 4 risques de prédictions. On donnera la courbe du risque de validation croisée en fonction du paramètre de régularisation (on veillera à bien choisir l’espace de définition pour le graphe). Donner le paramètre de régularisation optimal et la valeur du risque sur le test.

```python
alpha = range(0,100)
mse_ridge = []
for a in alpha:
    #initialiser risque de prédiction pour chaque alpha
    mse_tmp = []
    #parametrer alpha de Ridge
    ridge = Ridge(alpha=a)
    #pour les 4 folds
    for i in range(4):
        test = list(list_kf[i][1])
        train = list(list_kf[i][0])
        model = ridge.fit(x_train.iloc[train,:], y_train.iloc[train]) 
        ypred_ridge = model.predict(x_train.iloc[test,:])
        mse = ((y_train.iloc[test] - ypred_ridge)**2).mean()
        mse_tmp.append(mse)
    mse_tmp = pd.DataFrame(mse_tmp)
    mse_ridge.append(mse_tmp.mean())
mse_ridge = pd.DataFrame(mse_ridge)


#Index du min du ridge => optimalité 
i = int(mse_ridge.idxmin())
mse_ridge_min = float(mse_ridge.min())
alpha_optimal = alpha[i]
print("Paramètre optimal : ", alpha_optimal)

# Ridge optimal 

rdg_opt = Ridge(alpha=alpha_optimal).fit(x_train, y_train)
ypred_rdg_opt = rdg_opt.predict(x_test)
mse_rdg_opt = ((y_test - ypred_rdg_opt) ** 2).mean()

print("Risque de prédiction sur le ridge avec paramètre optimal : ", mse_rdg_opt)
```

```python
plt.figure()
plt.plot(alpha[30:100], mse_ridge[30:100])
plt.title("Risque de validation croisée en fonction du paramètre de régularisation")
plt.xlabel("Paramètre de régularisation")
plt.ylabel("Risque de validation croisée")
plt.axhline(mse_ridge_min, color='r', ls='--')
plt.axvline(alpha_optimal, color='r', ls='--')
plt.show()
```

## Question 13 

A l’aide de la fonction lassoCV de sklearn, choisir le paramètre de régularisation du LASSO. Donner le risque de prédiction associé.

```python
#Lasso Cross Validation
reg_lasso = LassoCV(random_state= 2).fit(x_train, y_train)

# Paramètre optimal de régularisation
print("Paramètre de régularisation :", reg_lasso.alpha_)

#Risque de prédiction 
ypred_lasso = reg_lasso.predict(x_test)
mse_lasso = ((ypred_lasso - y_test)**2).mean()
print("Risque de prédiction Lasso : ", mse_lasso)
```

## Question 14 

Donner les variables selectionées par le lasso. Combien y-en a t-il ? Appliquer la méthode OLS aux variables sélectionnées. Cette méthode est appelé Least-square LASSO.

```python
coef_lasso = reg_lasso.coef_
x_lasso = np.asarray(np.nonzero(coef_lasso))[0]
print(len(x_lasso), "variables sélectionnées par le Lasso", x_lasso)

# OLS sur variables sélectionnées 

ols_lasso = LinearRegression().fit(x_train.iloc[:,x_lasso], y_train)

#Risque de prédiction 
ypred_lslasso = ols_lasso.predict(x_test.iloc[:,x_lasso])
mse_lslasso = ((ypred_lslasso - y_test)**2).mean()
print("Risque de prédiction Least Square Lasso : ", mse_lslasso)
```

### Risques de prédiction sur toutes les méthodes 

```python
print("Risque de prédiction sur échantillon test en forward selection : {0}".format(mspe_fselec))
print("Risque de prédiction sur échantillon test OLS :", mse_ols)
print("Risque de prédiction sur échantillon test PCA before OLS:", mse_pca)
print("Risque de prédiction sur le ridge avec paramètre optimal : ", mse_rdg_opt)
print("Risque de prédiction Lasso : ", mse_lasso)
print("Risque de prédiction Least Square Lasso : ", mse_lslasso)

print("\nForward selection < Lasso < PCA before OLS < Ridge < LS Lasso < OLS")
```
